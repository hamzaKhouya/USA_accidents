<center><h1>Description</h1></center>
This is a countrywide car accident dataset, which covers 49 states of the USA. The accident data are collected from February 2016 to Dec 2021, using multiple APIs that provide streaming traffic incident (or event) data. These APIs broadcast traffic data captured by a variety of entities, such as the US and state departments of transportation, law enforcement agencies, traffic cameras, and traffic sensors within the road-networks. Currently, there are about 2.8 million accident records in this dataset. Check here to learn more about this dataset. 
You can download data here <a href="https://www.kaggle.com/datasets/sobhanmoosavi/us-accidents" target="_blank" >Dataset</a>

```python
import pandas as pd
import numpy as np
import missingno as msno

data = pd.read_csv("Data/US_Accidents_March23.csv")
print(data.shape)
data.columns
```

```python
data.head()
```

<h1>1. Data Preparation</h1>

```python
data.info()
```

## Data description

```python
data.describe().T
```

### Delete unused columns


```python
data.drop(columns=['Country','Timezone','Source',
                   'Description','Street','County','Zipcode','Airport_Code',
                   'End_Lat', 'End_Lng','Bump', 'Crossing', 'Give_Way', 
                   'Junction', 'No_Exit', 'Railway','Roundabout', 'Station', 
                   'Stop', 'Traffic_Calming', 'Traffic_Signal','Turning_Loop', 
                   'Civil_Twilight', 'Nautical_Twilight','Astronomical_Twilight'],
          inplace=True)
data.columns
```

# Rename columns

```python
data.rename(columns={'Start_Lat' : 'Latitude', 'Start_Lng' : 'Longitude'}, inplace = True)
```

## Convert Objects dates to Timestamp

```python
# this function convert date columns from Objects to timestamp
def convert_date_columns(columns):
    for column in columns:
        data[column] = pd.to_datetime(data[column])
    return data
convert_date_columns(['Start_Time','End_Time','Weather_Timestamp'])
data.info()
```

<h1 align='center'> Working with missing data </h1>

```python
null_cols = [i for i in data.columns if data[i].isnull().any()]
print(null_cols)
msno.matrix(data[null_cols]);
```

```python
missing_values = data.isna().sum().sort_values(ascending=False)
missing_percentage = missing_values[missing_values!=0]/len(data)*100
print(" Percentage of Missing Values \n", missing_percentage)
```

```python
data[['City', 'Weather_Timestamp', 'Temperature(F)', 'Wind_Chill(F)', 
      'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Direction',
      'Wind_Speed(mph)', 'Precipitation(in)', 'Weather_Condition',
      'Sunrise_Sunset']].head()
```

### Replace Numerical data by mean

```python
# Temperature
temp_mean = data['Temperature(F)'].mean()
print('Temperature mean : ' , temp_mean)
data['Temperature(F)'].replace(np.nan, temp_mean , inplace = True)

# Wind_Chill
wind_mean = data['Wind_Chill(F)'].mean()
print('Wind_Chill mean : ' ,wind_mean)
data['Wind_Chill(F)'].replace(np.nan, wind_mean , inplace = True)

# Humidity
humidity_mean = data['Humidity(%)'].mean()
print('Humidity mean : ' ,humidity_mean)
data['Humidity(%)'].replace(np.nan, humidity_mean , inplace = True)

# Pressure(in)
pressure_mean = data['Pressure(in)'].mean()
print('Pressure mean : ' ,pressure_mean)
data['Pressure(in)'].replace(np.nan, humidity_mean , inplace = True)

# Visibility
visibility_mean = data['Visibility(mi)'].mean()
print('Visibility mean : ' ,visibility_mean)
data['Visibility(mi)'].replace(np.nan, humidity_mean , inplace = True)

# Wind_Speed(mph)
wind_speed_mean = data['Wind_Speed(mph)'].mean()
print('Wind_Speed(mph) mean : ' ,wind_speed_mean)
data['Wind_Speed(mph)'].replace(np.nan, wind_speed_mean , inplace = True)

# Precipitation(in)
precipitation_mean = data['Precipitation(in)'].mean()
print('Precipitation(in) mean : ' ,precipitation_mean)
data['Precipitation(in)'].replace(np.nan, precipitation_mean , inplace = True)
```

### Replace non numerical data with most frequent value

```python
# Wind_direction
wind_direction_value = data['Wind_Direction'].value_counts().idxmax()
print('Wind_Direction frequent value : ' , wind_direction_value)
data['Wind_Direction'].replace(np.nan, wind_direction_value , inplace = True)

# Weather_Timestamp
weather_condition_value = data['Weather_Condition'].value_counts().idxmax()
print('Weather_Condition frequent value : ' , weather_condition_value)
data['Weather_Condition'].replace(np.nan, weather_condition_value , inplace = True)


# Sunrise_Sunset
sunrise_sunset_value = data['Sunrise_Sunset'].value_counts().idxmax()
print('Sunrise_Sunset frequent value : ' , sunrise_sunset_value)
data['Sunrise_Sunset'].replace(np.nan, sunrise_sunset_value , inplace = True)
```

### Delete null values for City and Weather_timestamp

```python
print('All data shape: ', data.shape)
print('Shape of data with null city or weather timestamp : ',
      data.loc[data['City'].isnull() | data['Weather_Timestamp'].isnull()].shape)
data.dropna(subset = ['City','Weather_Timestamp'], inplace = True)
print('All data shape: ', data.shape)
```
